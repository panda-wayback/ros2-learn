1.导入rclpy和Node

````python
import rclpy
from rclpy.node import Node
````

2.导入TF帧：TF帧对应的消息接口为geometry_msgs.msg下的TransformStamped
其结构可以使用下面指令查看：

```
!ros2 interface show geometry_msgs/msg/TransformStamped
```

````python
!ros2 interface show geometry_msgs/msg/TransformStamped
````

````python
from geometry_msgs.msg import TransformStamped
````

3.从tf2_ros包中导入静态坐标变换广播器

````python
from tf2_ros import StaticTransformBroadcaster
````

4.初始化节点

````python
rclpy.init()
node = Node("transform_node")
````

5.构造静态广播发布器

````python
static_tf_pub = StaticTransformBroadcaster(node)
````

6.构造TF帧，，其要发布出去的消息

````python
t = TransformStamped()
t.header.stamp = node.get_clock().now().to_msg()
# parent-name:基坐标系B
t.header.frame_id = 'B'
# chrild-name：相机坐标系C
t.child_frame_id = 'C'
# 平移关系，单位m
t.transform.translation.x = 0.0
t.transform.translation.y = 2.0
t.transform.translation.z = 0.0
# 旋转关系，四元数形式，我们需要将欧拉角的形式转换成四元数
# 可以使用在线坐标转换工具：https://quaternions.online/
t.transform.rotation.x = 0.0
t.transform.rotation.y = 0.0
t.transform.rotation.z = 0.0
t.transform.rotation.w = 1.0
````

7.发布坐标关系

````python
static_tf_pub.sendTransform(t)
````
